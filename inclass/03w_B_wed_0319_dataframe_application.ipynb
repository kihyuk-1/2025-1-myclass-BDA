## df.query()

```python
df = pd.DataFrame({
    'A': range(1, 6),
    'B': range(5, -4, -2),  # range(5, -4, -2)로 수정
    'C': range(0, 5),
    'D clmns ': range(3, -2, -1),
    'E': ['python', 'java', 'python', 'go', 'python']
})
```

```python
df.query('A > B')
```

```python
df[df.A > df.B]
```

```python
df.query('C == 10')
```

```python
df.query('A > 0')
```

```python
df.query('E == "python"')  # 또는 df.query('E == \'python\'')
```

```python
df.query('E !="python"')
```

```python
df.query("A == 'Dlmns'")
```
